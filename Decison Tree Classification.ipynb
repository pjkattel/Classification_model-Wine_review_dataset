Convert the categorical variable into ohe encoded variable. The description contions some key words that describes the quality of the wine. We need a function to read the line and capture the words to get the columns of each features.

```python
def conditional_column(dataframe, column, keyword):
    value1 = []
    for index,row in dataframe.iterrows():
        if keyword in row[column]:
            value = 1
        else:
            value = 0
        value1.append(value)
    return value1
    
```

```python
essential_wine_df['vanilla'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="vanilla")
essential_wine_df['currant'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="currant")
essential_wine_df['oak'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="oak")
essential_wine_df['complex'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="complex")
essential_wine_df['buttery'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="buttery")
essential_wine_df['herb'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="herb")
essential_wine_df['orange'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="orange")
essential_wine_df['flower'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="flower")
essential_wine_df['crisp'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="crisp")
essential_wine_df['toast'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="toast")
essential_wine_df['pineapple'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="pineapple")
essential_wine_df['peach'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="peach")
essential_wine_df['classic'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="classic")
essential_wine_df['noble'] = conditional_column(dataframe=essential_wine_df, column="description", keyword="noble")

# Print the top 5 rows of data from the essential_wine_df dataframe
essential_wine_df.head(5)
```

Splitting the data into training and test sets 

```python
# Create a Numpy array for every column that is not 'title_encoded' (features)
# Assign the resulting matrix into a variable named "X"
X = essential_wine_df.iloc[:,1:].values

# Print the top 5 rows of data from the X dataframe
print(X[0:5,:])

# Create a Numpy array for the column 'title_encoded' (label)
# Assign the resulting matrix into a variable named "y"
y = essential_wine_df.iloc[:,0].values

# Print the top 5 rows of data from the y dataframe
print(y[0:5])
```

```python
# Train, Test, and Split your Numpy array data sets (features and label)
# Used a test size of 0.001
# Used a random state of 42
# Assigned the resulting matrix into four variables named "X_train, X_test, y_train, y_test"
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.001,random_state = 42)

print(X_train, X_test, y_train, y_test)
```

```python
# Instantiate the Decision Tree Classifier Model
# Use a random state of 42
# Use a max depth of 10
# Assign the resulting matrix into four variables named "model"
model = DecisionTreeClassifier(random_state = 42, max_depth =10)

# TODO: Print the Model Statistics
print(model)

# TODO: Fit your Training Data to the Decision Tree Classifier Model
# Assign the resulting matrix into four variables named "model"
model = model.fit(X_train,y_train)

def predict_top_five_wines(feature_preds):
    top_five_pred_list = []
    number_one_pred = model.predict(feature_preds)
    number_one_pred_name = label_encoder.inverse_transform(number_one_pred)
    top_five_pred_list.append(number_one_pred_name[0])
    top_four = np.argsort(model.predict_proba(feature_preds), axis=1)[0][-4:][::-1]
    for each_wine in top_four:
        top_four_wine_name = label_encoder.inverse_transform(each_wine)
        top_five_pred_list.append(top_four_wine_name)
    return top_five_pred_list

class BeOurGuest:
    def __init__(self, points, price, vanilla, currant, oak, complex, buttery, herb, orange, flower, crisp, toast,
                 pineapple, peach, classic, noble, argentina, australia, austria, brazil, bulgaria, canada, chile,
                 croatia, england, france, germany, greece, hungary, israel, italy, lebanon, macedonia, mexico,
                 moldova, morocco, new_zealand, peru, portugal, romania, slovenia, south_africa, spain,
                 switzerland, turkey, us, ukraine, uruguay, cabernet_sauvignon, chardonnay, merlot, pinot_noir,
                 white_riesling):
        self.points = points
        self.price = price
        self.vanilla = vanilla
        self.currant = currant
        self.oak = oak
        self.complex = complex
        self.buttery = buttery
        self.herb = herb
        self.orange = orange
        self.flower = flower
        self.crisp = crisp
        self.toast = toast
        self.pineapple = pineapple
        self.peach = peach
        self.classic = classic
        self.noble = noble
        self.argentina = argentina
        self.australia = australia
        self.austria = austria
        self.brazil = brazil
        self.bulgaria = bulgaria
        self.canada = canada
        self.chile = chile
        self.croatia = croatia
        self.england = england
        self.france = france
        self.germany = germany
        self.greece = greece
        self.hungary = hungary
        self.israel = israel
        self.italy = italy
        self.lebanon = lebanon
        self.macedonia = macedonia
        self.mexico = mexico
        self.moldova = moldova
        self.morocco = morocco
        self.new_zealand = new_zealand
        self.peru = peru
        self.portugal = portugal
        self.romania = romania
        self.slovenia = slovenia
        self.south_africa = south_africa
        self.spain = spain
        self.switzerland = switzerland
        self.turkey = turkey
        self.us = us
        self.ukraine = ukraine
        self.uruguay = uruguay
        self.cabernet_sauvignon = cabernet_sauvignon
        self.chardonnay = chardonnay
        self.merlot = merlot
        self.pinot_noir = pinot_noir
        self.white_riesling = white_riesling

    def wine_pref_params(self):
        return [[self.points, self.price, self.vanilla, self.currant, self.oak, self.complex, self.buttery, self.herb, self.orange,
                 self.flower, self.crisp, self.toast, self.pineapple, self.peach, self.classic, self.noble,
                 self.argentina, self.australia, self.austria, self.brazil, self.bulgaria, self.canada, self.chile, self.croatia,
                 self.england, self.france, self.germany, self.greece, self.hungary, self.israel, self.italy, self.lebanon, self.macedonia,
                 self.mexico, self.moldova, self.morocco, self.new_zealand, self.peru, self.portugal, self.romania, self.slovenia, self.south_africa,
                 self.spain, self.switzerland, self.turkey, self.us, self.ukraine, self.uruguay, self.cabernet_sauvignon, self.chardonnay,
                 self.merlot, self.pinot_noir, self.white_riesling ]]
```
